```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt


# f(x) = 1/sqrt(x) x E [1;10]

x = np.linspace(1,10,200).reshape(-1, 1)

X = np.hstack([x, np.sqrt(x), np.log(x), 1 / x, x**2])

f = 1 / np.sqrt(x)
y = np.hstack([f, f**2, np.sqrt(f)])

model = Sequential(
    [
        Dense(64, activation="relu", input_shape=(5,)),
        Dense(32, activation="relu"),
        Dense(3), 
    ]
)

model.compile(optimizer="adam", loss="mse")
model.fit(X, y, epochs=300, batch_size=32, verbose=1)

y_pred = model.predict(X)

plt.figure(figsize=(7, 6))
plt.scatter(y[:, 0], y_pred[:, 0], color="blue", alpha=0.6)
plt.plot([0, 1], [0, 1], "r--", lw=2)
plt.title("Przewidywanie f(x)=1/sqrt(x)")
plt.xlabel("Prawdziwa wartość (f(x))")
plt.ylabel("Wartość przywidywana")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt

# f(x) = 1/sqrt(x) x E [1;10]

X = np.linspace(1, 10,).reshape(-1, 1)
y = 1 / np.sqrt(X)

model = Sequential(
    [
        Dense(64, activation="relu", input_shape=(1,)),
        Dense(32, activation="relu"),
        Dense(1), 
    ]
)

model.compile(optimizer="adam", loss="mse")
history = model.fit(X, y, epochs=131, batch_size=32, verbose=1)

y_pred = model.predict(X)

plt.figure(figsize=(7, 6))
plt.scatter(y, y_pred, color="blue", alpha=0.6)
plt.plot([0, 1], [0, 1], "r--", lw=2)
plt.title("Regresja f(x)=1/sqrt(x)")
plt.xlabel("Wartość rzeczywista f(x)")
plt.ylabel("Wartość przewidywana")
plt.grid(True)
plt.show()

mse_fn = tf.keras.losses.MeanSquaredError()
mse = mse_fn(y, y_pred).numpy()

print(f"\nMean Squared Error: {mse:.5f}")
```
